# Association Rule Using Apriori Algorithm
Here in this notebook we have taken a dataset containing 7501 rows each row contain a transaction each column represent a particular
item in transaction then we applied apriori alogrithm to find association between item this dataset is  belongs to a supermarket from france 

Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

importing dataset using read_csv method from pandas

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
```

Have a look at dataset here we can see each column contains a items in transaction and each row represnts a single transaction

```python
dataset.head()
```

Here we have converted each row into vector and added into transactions list making it matrix 

transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])

Here we have imported module and create apriori model with minimium support of 0.003 and minimium confidence of 0.2 and minimium lift of 3 and minimiun length of two

```python
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

here we have created rules and printed all rules with above specifications

```python
results = list(rules)
for i in results:
    print(i)
```

from above solution it can be interpreted that the first rule is light cream with chicken it has confidence of 29% and second 
rule is esclope with mushroom cream sauce with confidence of 30% likewise we can interpret all rules although first five to ten rules of more significance as there are more chance of them to occur
